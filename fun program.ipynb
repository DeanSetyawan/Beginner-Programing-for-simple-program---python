```python
A=input("masukan angka pertama= ")
B=input("Masukan angka kedua= ")
C=input("masukan tipe operator= ")
D=int(A)
E=int(B)
if(C=="Penjumlahan"):
  F=D+E
  print("hasil penjumlahan=",F)
if(C=="Pengurangan"):
  F=D-E
  print("Hasil pengurangan=",F)
if(C=="Perkalian"):
  F=D*E
  print("hasil perkalian=",F)
if(C=="Pembagian"):
  F=D/E
  print("hasil pembagian=",F)
if(C=="pangkat"):
  F=D**E
  print("Hasil angka pertama pangkat angka kedua =",F)
if(C=="akar"):
  F=D**E
  print("Hasil angka pertama pangkat angka kedua =",F)
```
